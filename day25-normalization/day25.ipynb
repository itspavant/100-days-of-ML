# Normalisation

Normalization is a technique often applied as apart of data preparation for Machine Learning. The goal of Normalisation is to change the values of Numerical columns in the dataset to use a common scale, without distorting differences in the ranges of values or losing information.

`- Eliminate units and keep magnitudes on Normal Scale`

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class label', axis=1),
                                                   df['Class label'],
                                                   test_size = 0.3,
                                                   random_state = 0)
X_train.shape, X_test.shape
```

## Min-Max Normalization

Min-Max Normalization is trying to scale down the date in to the range [0, 1]



`it's like compressing/squashing the spread of data into a unit cube`   



after min-max normalization, the data is fitted into the range of [0, 1]. but we cannot comment anything about how mean and std deviation are affected.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Alcohol'], X_train['Malic acid'],c=y_train)
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Alcohol'], X_train_scaled['Malic acid'],c=y_train)
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Alcohol'], ax=ax1)
sns.kdeplot(X_train['Malic acid'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Alcohol'], ax=ax2)
sns.kdeplot(X_train_scaled['Malic acid'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Alcohol Distribution Before Scaling')
sns.kdeplot(X_train['Alcohol'], ax=ax1)

# after scaling
ax2.set_title('Alcohol Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Alcohol'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Malic acid Distribution Before Scaling')
sns.kdeplot(X_train['Malic acid'], ax=ax1)

# after scaling
ax2.set_title('Malic acid Distribution After Standard Scaling')
sns.kdeplot(X_train_scaled['Malic acid'], ax=ax2)
plt.show()
```

`The distribution curve needn't be of same shape before and after scaling, it can vary`

outliers affect the scaling as they push themselves into the unit cubicle making the data become too compressed, though there is innefective outliers impact on target.

## Mean Normalization 



- It does `mean centering`
- Scales b/w [-1, 1]
- Useful in Algorithms where centered data is required. but in such cases, standardisation is prefferred.
- In Sci-Kit Learn, no class is available for Mean Normalization.

## MaxAbs Scaling



- MaxAbsScaler class available in sci-kit learn
- used in cases where very sparse data is available, i.e. more zeroes are present

## Robust Scaling



- IQR = 75th percentile - 25th percentile
- Robust to outliers

## Normalization vs Standardization

1. Is Feature scaling required?
2. Mostly Standardisation is used, but MinMax is used in cases where Min and Max is well known before only. as in the cases of CNN.

Tips to decide which scaling to use:
- if know before only about min and max, i.e range of the numerical column is known to be in [min, max] -> use MinMax Normalization
- if No idea what to do -> Standardization
- if outliers are present extensively -> Robust Scaling
- if sparse matrices are presnt extensively -> MaxAbs Scaling
