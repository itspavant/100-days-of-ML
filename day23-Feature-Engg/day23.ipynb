# Feature Engineering

Feature Engineering is the process of using domain knowledge to extract features form raw data. These features can be used to improve the performance of Machine Learning algorithm

# Types of Feature Engineering

Feature Engineering can be said to have the following subsets



# 1. Feature Transformation

Feature transformation is the process of the converting an existing column to some other different type so the performance of the Machine Learning Algorithm can be improved and enhanced

## 1.1 Missing value imputation

Missing values are very common and one of the painful scenarios in real world datasets. missing values arise due to various number of reasons, such as user incomplete data submission, or inefficient data collection, etc.

and these empty values decrease the efficiency of an algorithm and can become unacceptable by modules like Sci-kit learn, and hence it is very important to deal with the missing values either by removing them or by filling them.

Handling missing values becomes the first step in the pipeline of feature engg.

# 1.2 Handling Categorical Features



categorical data can be not accepted by Sci-Kit Learn and also the numerical values enhance the performance of machine learning model. and hence numerical values are preffered over categorical.

with techniques like `one hot encoding` , etc. categorical features can be handled

# 1.3 Outliers Detection



outliers can significantly affect the performance, prediction and efficiency of an algorithm by completing changing the decisiveness. so it becomes extremely important to detect and remove the outliers.

# 1.4 Feature Scaling



- Feature scaling is the process of bringing all numerical features into a similar range so that their magnitudes do not distort a models learning.
- It is important because features with large values can dominate those with smaller values, especially in distance-based algorithms.
- so feature scaling is necessary to be done by techniques like, `min-max normalisation` `z-score normalisation` , etc

# 2. Feature Construction



Feature Construction is using the domain knowledge and with the deeper understading about the dataset, we can come up with new features that can have more meaningful impact on the predictions


##

# 3. Feature Selection



Feature selection is the process of choosing only the most relevant and useful features from a dataset to build a model and disregarding the features that have very slight to zero impact on the Algorithm and prediction

##

# 4. Feature Extraction



Feature extraction is programmatically (not manually) creating new features and reducing redundant features so the efficiency of the algorithm can be improved.

completely new features are created, and reduced also through methods like `PCA` , `LCA` - will be discussed in detail in later classes
